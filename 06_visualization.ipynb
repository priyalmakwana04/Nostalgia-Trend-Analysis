```python
import pandas as pd
import matplotlib.pyplot as plt
import os

path = "../data/final_dataset.csv"   # ✅ this is what we saved in feature engineering
print("File exists?", os.path.exists(path))

df = pd.read_csv(path)
print("Loaded shape:", df.shape)
print("Columns:", df.columns.tolist())

df.head()
```

```python
# if year is not numeric, fix it
df["year"] = pd.to_numeric(df["year"], errors="coerce")

yearly = df.groupby("year")["engagement_score"].mean()

print("Yearly series preview:")
print(yearly.head())

yearly.plot(kind="line")
plt.title("Average Engagement Over Time")
plt.xlabel("Year")
plt.ylabel("Avg Engagement Score")
plt.show()
```

```python
print("Revival label counts:")
print(df["revival_label"].value_counts())

df["revival_label"].value_counts().plot(kind="bar")
plt.title("Revival Label Count")
plt.xlabel("Label (0=No Revival, 1=Revival)")
plt.ylabel("Count")
plt.show()
```
