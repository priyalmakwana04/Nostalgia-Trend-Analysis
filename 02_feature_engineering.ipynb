```python
import pandas as pd
import numpy as np

# Clean column names
df.columns = df.columns.str.strip().str.lower()

print("✅ Columns found:", df.columns.tolist())

# ---- AUTO FIND HASHTAG COLUMN ----
possible_hashtag_cols = [
    "hashtag", "hashtags", "tag", "tags", "trend", "trends", "keyword", "keywords", "topic"
]

hashtag_col = None
for c in possible_hashtag_cols:
    if c in df.columns:
        hashtag_col = c
        break

if hashtag_col is None:
    # try partial match (e.g., "hash_tag", "hashtag_name")
    for c in df.columns:
        if "hash" in c or "tag" in c or "trend" in c or "keyword" in c:
            hashtag_col = c
            break

if hashtag_col is None:
    raise ValueError("❌ Could not find a hashtag/trend column. Paste df.columns.tolist() here.")

df["hashtag"] = df[hashtag_col].astype(str).str.strip().str.lower()
print("✅ Using hashtag column:", hashtag_col)


# ---- AUTO CREATE YEAR (and month if possible) ----
# 1) If year column exists with any name
possible_year_cols = ["year", "post_year", "created_year", "publish_year"]
year_col = None

for c in possible_year_cols:
    if c in df.columns:
        year_col = c
        break

# 2) Else find date column and extract year/month
possible_date_cols = ["date", "post_date", "created_at", "timestamp", "time", "published_at"]

date_col = None
for c in possible_date_cols:
    if c in df.columns:
        date_col = c
        break

if year_col is not None:
    df["year"] = pd.to_numeric(df[year_col], errors="coerce")
    print("✅ Using year column:", year_col)

elif date_col is not None:
    df[date_col] = pd.to_datetime(df[date_col], errors="coerce")
    df["year"] = df[date_col].dt.year
    df["month"] = df[date_col].dt.month
    print("✅ Created year/month from date column:", date_col)

else:
    # If no year and no date, we create a dummy year to allow project to continue
    df["year"] = 2026
    df["month"] = 1
    print("⚠️ No year/date column found. Created dummy year=2026, month=1 (project will still run).")


print("\n✅ Final check:")
print("hashtag exists?", "hashtag" in df.columns)
print("year exists?", "year" in df.columns)
df[["hashtag", "year"] + (["month"] if "month" in df.columns else [])].head()
```

```python
# Engagement score (use whatever exists)
possible = ["likes", "comments", "shares", "views"]
available = [c for c in possible if c in df.columns]
for c in available:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

df["engagement_score"] = df[available].sum(axis=1)

# Hashtag frequency
hashtag_year = df.groupby(["hashtag","year"]).size().reset_index(name="trend_count")
hashtag_year.head()
```

```python
possible = ["likes", "comments", "shares", "views"]
available = [c for c in possible if c in df.columns]

print("Engagement columns found:", available)

for c in available:
    df[c] = pd.to_numeric(df[c], errors="coerce").fillna(0)

df["engagement_score"] = df[available].sum(axis=1)
df[available + ["engagement_score"]].head()
```

```python
hashtag_year = df.groupby(["hashtag","year"]).size().reset_index(name="trend_count")
hashtag_year.head(10)
```

```python
sort_cols = ["hashtag", "year"]
if "month" in df.columns:
    sort_cols.append("month")

df = df.sort_values(sort_cols).reset_index(drop=True)

df["engagement_growth"] = df.groupby("hashtag")["engagement_score"].pct_change()
df[["hashtag","year"] + (["month"] if "month" in df.columns else []) + ["engagement_score","engagement_growth"]].head(15)
```

```python
threshold = 1.5   # 150% growth
df["revival_label"] = df["engagement_growth"].apply(lambda x: 1 if pd.notnull(x) and x > threshold else 0)

df["revival_label"].value_counts()
```

```python
out_path = "../data/final_dataset.csv"
df.to_csv(out_path, index=False)

print("Saved:", out_path)
print("File exists now?", os.path.exists(out_path))
```
